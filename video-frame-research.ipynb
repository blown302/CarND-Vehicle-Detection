```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
frame_no = 0 * 25

cap = cv2.VideoCapture('white_car_dist.mp4')
cap.set(1, frame_no)

ret, frame = cap.read() # Read the frame
image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(image)

plt.imsave('test_images/white_car_from_video.jpg', image)
```
